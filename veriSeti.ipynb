```python
import pandas as pd
import matplotlib.pyplot as plt

veriSeti = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/veri.csv/data.csv')

turkce_sutun_isimleri = ['Id', 'Dizi Adı', 'Türü', 'IMBD', 'num', 'Yıl']
veriSeti.columns = turkce_sutun_isimleri

print(veriSeti.head(20).to_string())

veriSeti = veriSeti.sort_values(by=['IMBD'], ascending=False).head(20)

diziler = veriSeti['Dizi Adı']
imbd = veriSeti['IMBD']

plt.figure(figsize=(20, 7))
plt.bar(diziler, imbd)
plt.xlabel('Diziler')
plt.ylabel('IMBD Puan')

plt.title('IMBD si en yüksek 20 dizi')
plt.xticks(rotation=90)
plt.ylim(9, 9.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

veriSeti = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/veri.csv/data.csv')

turkce_sutun_isimleri = ['Id', 'Dizi Adı', 'Türü', 'IMBD', 'num', 'Yıl']
veriSeti.columns = turkce_sutun_isimleri

veriSeti = veriSeti.sort_values(by=['IMBD'], ascending=False).head(20)

genre_ratings = veriSeti.groupby('Türü')['IMBD'].mean().sort_values(ascending=False)

diziler = veriSeti['Dizi Adı']
yil = veriSeti['Türü']

plt.figure(figsize=(20, 7))
plt.bar(diziler, yil)
plt.xlabel('Diziler')
plt.ylabel('Türleri')

plt.title('IMBD Oranı Yüksek 20 Dizinin Türlere Göre Dağılımı')
plt.xticks(rotation=90)
plt.show()
```
